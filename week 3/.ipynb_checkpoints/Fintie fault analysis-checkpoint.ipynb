```python
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Folder path containing .fsp files
folder_path = r"C:\Users\ASUS\OneDrive\Desktop\SDA\week 3\srcmod2024-12-02FSP"

# List all .fsp files in the folder
fsp_files = [f for f in os.listdir(folder_path) if f.endswith(".fsp")]

# Read each file and generate plots
for file_name in fsp_files:
    file_path = os.path.join(folder_path, file_name)
    
    # Read and extract numerical data
    numerical_data = []
    with open(file_path, "r") as file:
        for line in file:
            if line.startswith("%"):
                continue  # Skip comments
            values = line.split()
            try:
                numerical_values = list(map(float, values))  # Convert valid numerical lines
                numerical_data.append(numerical_values)
            except ValueError:
                continue  # Skip non-numeric lines

    # Convert to numpy array and process if data exists
    if not numerical_data:
        print(f"Skipping {file_name}, no valid numerical data found.")
        continue
    
    numerical_data = np.array(numerical_data)

    # Extract relevant columns
    lat = numerical_data[:, 0]
    lon = numerical_data[:, 1]
    slip_strike = numerical_data[:, 2]
    slip_dip = numerical_data[:, 3]

    # Create 2D contour plots
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # 2D Contour plot for Slip Along Strike
    contour1 = axes[0].tricontourf(lon, lat, slip_strike, levels=20, cmap="viridis")
    plt.colorbar(contour1, ax=axes[0])
    axes[0].set_title(f"2D Contour (Strike) - {file_name}")
    axes[0].set_xlabel("Longitude")
    axes[0].set_ylabel("Latitude")

    # 2D Contour plot for Slip Along Dip
    contour2 = axes[1].tricontourf(lon, lat, slip_dip, levels=20, cmap="plasma")
    plt.colorbar(contour2, ax=axes[1])
    axes[1].set_title(f"2D Contour (Dip) - {file_name}")
    axes[1].set_xlabel("Longitude")
    axes[1].set_ylabel("Latitude")

    plt.tight_layout()
    plt.savefig(os.path.join(folder_path, f"{file_name}_2D.png"))
    plt.close()

    # Create 3D Contour plots
    fig3d = plt.figure(figsize=(8, 6))
    ax3d = fig3d.add_subplot(111, projection="3d")
    ax3d.plot_trisurf(lon, lat, slip_strike, cmap="viridis")
    ax3d.set_title(f"3D Contour (Strike) - {file_name}")
    ax3d.set_xlabel("Longitude")
    ax3d.set_ylabel("Latitude")
    ax3d.set_zlabel("Slip (Strike)")
    plt.savefig(os.path.join(folder_path, f"{file_name}_3D_Strike.png"))
    plt.close()

    fig3d2 = plt.figure(figsize=(8, 6))
    ax3d2 = fig3d2.add_subplot(111, projection="3d")
    ax3d2.plot_trisurf(lon, lat, slip_dip, cmap="plasma")
    ax3d2.set_title(f"3D Contour (Dip) - {file_name}")
    ax3d2.set_xlabel("Longitude")
    ax3d2.set_ylabel("Latitude")
    ax3d2.set_zlabel("Slip (Dip)")
    plt.savefig(os.path.join("C:\Users\ASUS\OneDrive\Desktop\SDA\week 3", f"{file_name}_3D_Dip.png"))
    plt.close()

    print(f"Plots saved for {file_name}")

print("All files processed successfully.")
```
